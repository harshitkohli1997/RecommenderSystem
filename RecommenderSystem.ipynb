```python
import numpy as np 
import pandas as pd
```

```python
columns_name = ['user_id','item_id','rating','timestamp']
```

```python

df = pd.read_csv('u.data',sep = '\t',names=columns_name)
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python

df = pd.merge(df,movies_titles,on='item_id')
df.head()
```
